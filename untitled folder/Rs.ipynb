### How the following Class works is:
- Step 1: User passes two values to the class: EventType and city. EventType is the type of event that he wants to conduct( Birthday, Marriage etc).

- Step 2: Then the EventType and city values are compared with Dataframe df and extracted the rows containing the EventType and City. For eg. if a user passes Birthday and Sarajevo to the class then those rows containing Birthday as Event_type and city are stored in result dataframe

    - Popularity algorithm works by calculating average Customer rating of a hotel for a particular event type and arranges it in descending order. Those hotels with highest popularity are returned as most popular now.

- Step 3: For this we find the average rating of a hotel for a particular event type and arrange it in descending order and store it in avg dataframe.
- Step 4: Initialize a new list -> new_list to store the names of hotels extracted from avg dataframe

- Step 5:Initialize another new list -> final_list to store the rows of the hotel stored in result dataframe that has hotel names 
stored in new_list

- Step 6: Return the row data stored in final_list as json format
